```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a large sample dataset
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=100000, freq='T')  # 100,000 minutes
data = np.random.randn(100000)  # Random data
df = pd.DataFrame({'Date': dates, 'Value': data})
df.set_index('Date', inplace=True)

# Define window size
window_size = 60  # 60 minutes

# Calculate rolling average and standard deviation using Pandas
df['Rolling_Avg'] = df['Value'].rolling(window=window_size).mean()
df['Rolling_Std'] = df['Value'].rolling(window=window_size).std()

# Optimizations
# 1. Use NumPy arrays for calculations to reduce overhead
values = df['Value'].values

# Rolling average
rolling_avg = np.convolve(values, np.ones(window_size) / window_size, mode='valid')

# Rolling standard deviation
def rolling_std(values, window_size):
    cumsum = np.cumsum(np.insert(values, 0, 0))
    cumsum2 = np.cumsum(np.insert(values**2, 0, 0))
    sum_ = cumsum[window_size:] - cumsum[:-window_size]
    sum2 = cumsum2[window_size:] - cumsum2[:-window_size]
    return np.sqrt((sum2 - sum_ ** 2 / window_size) / window_size)

rolling_std_dev = rolling_std(values, window_size)

# Ensure lengths match by padding the beginning with NaNs
rolling_avg_padded = np.concatenate([np.full(window_size - 1, np.nan), rolling_avg])
rolling_std_dev_padded = np.concatenate([np.full(window_size - 1, np.nan), rolling_std_dev])

# Append results to DataFrame
df['Rolling_Avg_Optimized'] = rolling_avg_padded
df['Rolling_Std_Optimized'] = rolling_std_dev_padded

# Plotting
plt.figure(figsize=(14, 7))
plt.plot(df.index, df['Value'], label='Original Data', alpha=0.5)
plt.plot(df.index, df['Rolling_Avg_Optimized'], label='Rolling Average (Optimized)', color='orange')
plt.plot(df.index, df['Rolling_Std_Optimized'], label='Rolling Std Dev (Optimized)', color='red')
plt.legend()
plt.title('Rolling Average and Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.show()
```
